```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

time1 = np.linspace(0,16.01)
orng1 = 20*np.ones(len(time1))

m = 30
x = 16
y = 20
b = 20-(x*m)

time2 = np.linspace(15.99,18)
orng2 = m*time2+b
line_color = 'orange'

plt.xkcd()
plt.plot(time1,orng1,color = line_color)
plt.xkcd()
plt.plot(time2,orng2,color = line_color)
plt.xkcd()
plt.xlabel('Time Since I was Born')
plt.ylabel('My Love for Oranges')
plt.xkcd()
plt.title("How You've Affected My Life")
plt.annotate('The Day I \nMet You', 
             xy = (16,20),arrowprops = dict(arrowstyle='->'),
            xytext = (10,40))
plt.yticks([])
plt.xticks([])
plt.xlim((0,18))
plt.ylim((0,100))
plt.savefig('oranges.png',bbox_inches = 'tight',pad_inches = 0.2)
```

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x_min = 0
x_max = 2
num_bars = 2
bar_coord = np.linspace(x_min,x_max,num_bars+2)[1:3]
bar_height = np.array([0, 100])

plt.xkcd()

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.xkcd()
ax.bar(bar_coord,bar_height,abs(x_max-x_min)/6,color = 'blue')
ax.set_xlim([x_min, x_max])
ax.set_ylim((0,max(bar_height)+10))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_xticks(bar_coord)
ax.set_xticklabels(['days i dont want\nto hug you','days i want\nto hug you'],fontsize = 10)

plt.title('just so you know')
for tic in ax.xaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False
ax.set_yticks(np.linspace(0,100,2))
plt.xkcd()
ax.set_yticklabels(['Literally not\none day','Alot of them'])
for tic in ax.yaxis.get_major_ticks():
    tic.tick1On = tic.tick2On = False
plt.savefig('hugs.png',bbox_inches = 'tight',pad_inches = 0.2)
```
